```python
import glob
import os

all_files = glob.glob(os.path.join('data/', "*.csv"))

df_full = pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
```

```python
descriptions = df_full['Job Description']
```

```python
from bs4 import BeautifulSoup

lists = []

for desc in descriptions[:2]:
    soup = BeautifulSoup(desc)
    # result = soup.find("a").previousSibling
    uls = soup.findAll('ul')
    lis = []
    for ul in uls:
        # print(ul)
        new = []
        for li in ul.findAll('li'):
            if li.find('ul'):
                break
            new.append(li.text.encode("utf-8"))
        lis.append(new)
    for l in lis:
        print(l)
    print('\n')

    # for li in lis:
    #     print li.text.encode("utf-8")
    # print(result.decode('utf-8'))
```
